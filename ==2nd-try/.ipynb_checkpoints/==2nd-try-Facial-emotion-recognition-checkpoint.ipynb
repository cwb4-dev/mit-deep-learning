# Facial Emotion Recognition

## **Problem Definition**

###**The context:** Why is this problem important to solve?<br>
* Machine facial expression recognition (FER) can help us to understand and respond to the emotions of others.
* FER can be used to detect and recognize emotions such as happiness, sadness, anger, fear, and surprise.
* It can be used, for example, to
    * improve human-computer interaction,
    * security and surveillance, and
    * medical diagnosis and treatment.  

### **The objectives:** What is the intended goal?<br>
The goal of this project is to use Machine Learning (specifically neural networks) to predict humman facial emotions.
###**The key questions:** What are the key questions that need to be answered?<br>
* Which FER CNN configuration most accurate predicting emotion in jpeg

###**The problem formulation:** What are we trying to solve using data science?
Correctly identify the emotion being displayed in a 48x48 jpeg picture.



## **About the dataset**

The data set consists of 3 folders, i.e., 'test', 'train', and 'validation'.
Each of these folders has four subfolders:

**‘happy’**: Images of people who have happy facial expressions.<br>
**‘sad’**: Images of people with sad or upset facial expressions.<br>
**‘surprise’**: Images of people who have shocked or surprised facial expressions.<br>
**‘neutral’**: Images of people showing no prominent emotion in their facial expression at all.<br>

## Model Evaluation Routines

```python
def model_data(model, train_set, validation_set, test_set):
    print("=== Train/Validation Data Evaluation ==============================")
    train_loss, train_accu = model.evaluate(train_set)
    test_loss, test_accu = model.evaluate(validation_set)
    print("final train accuracy = {:.2f} , validation accuracy = {:.2f}".format(train_accu*100, test_accu*100))
    print("=== Test Data Evaluation ==========================================")
    test_images, test_labels = next(test_set)
    accuracy = model.evaluate(test_images, test_labels, verbose = 2)
    return

def plot_loss_accuracy(hist):
    plt.figure(figsize=(14,5))
    plt.subplot(1,2,2)
    plt.plot(hist.history['accuracy'])
    plt.plot(hist.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(['train', 'test'], loc='upper left')

    plt.subplot(1,2,1)
    plt.plot(hist.history['loss'])
    plt.plot(hist.history['val_loss'])
    plt.title('Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    return

def plot_accuracy_loss(history):
    import seaborn as sns
    from matplotlib import pyplot

    sns.set()
    fig = pyplot.figure(0, (12, 4))
    pyplot.suptitle('Optimizer : Adam', fontsize=10)

    ax = pyplot.subplot(1, 2, 1)
    sns.lineplot(x=history.epoch, y=history.history['accuracy'], label='train')
    sns.lineplot(x=history.epoch, y=history.history['val_accuracy'], label='valid')
    pyplot.ylabel('Accuracy')
    pyplot.title('Model Accuracy')
    pyplot.tight_layout()

    ax = pyplot.subplot(1, 2, 2)
    sns.lineplot(x=history.epoch, y=history.history['loss'], label='train')
    sns.lineplot(x=history.epoch, y=history.history['val_loss'], label='valid')
    pyplot.ylabel('Loss')
    pyplot.title('Model Loss')
    pyplot.tight_layout()

    pyplot.show()
    return

def plot_confusion_matrix(test_set):
    from sklearn.metrics import classification_report
    from sklearn.metrics import confusion_matrix

    test_images, test_labels = next(test_set)
    pred = test_set.predict(test_images)
    pred = np.argmax(pred, axis = 1)
    y_true = np.argmax(test_labels, axis = 1)

    # Plotting the heatmap using confusion matrix
    cm = confusion_matrix(y_true, pred)
    plt.figure(figsize = (8, 5))
    sns.heatmap(cm, annot = True,  fmt = '.0f',
            xticklabels = data_classes,
            yticklabels = data_classes)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    return
```

## Create Image Generators for Model1 & Model2

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                #    rotation_range = 15,
                                #    brightness_range = (0., 2.),
                                #    width_shift_range = 0.15,
                                #    height_shift_range = 0.15,
                                #    shear_range = 0.15,
                                   zoom_range = 0.15)

training_set = train_datagen.flow_from_directory(
                                    train_dir,
                                    batch_size=batch_size,
                                    target_size=(image_size, image_size),
                                    shuffle=True,
                                    color_mode='grayscale',
                                    classes = data_classes,
                                    class_mode='categorical')

val_datagen = ImageDataGenerator(rescale=1./255,
                                #    rotation_range = 15,
                                #    brightness_range = (0., 2.),
                                #    width_shift_range = 0.15,
                                #    height_shift_range = 0.15,
                                #    shear_range = 0.15,
                                   zoom_range = 0.15)

val_set = val_datagen.flow_from_directory(
                                    val_dir,
                                    batch_size=batch_size,
                                    target_size=(image_size, image_size),
                                    shuffle=True,
                                    color_mode='grayscale',
                                    classes = data_classes,
                                    class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
                                    test_dir,
                                    target_size = (image_size, image_size),
                                    color_mode = "grayscale",
                                    batch_size = batch_size,
                                    classes = data_classes,
                                    class_mode = 'categorical',
                                    shuffle=True)




steps_per_epoch = training_set.n // training_set.batch_size
validation_steps = val_set.n // val_set.batch_size

ic(training_set.class_indices)
ic(steps_per_epoch)
ic(validation_steps);
```

## Model1

### **Define/Compile Model1**

```python
def model_1(input_size, classes=4):

    model = keras.Sequential()

    model.add(Conv2D(32, kernel_size=(3, 3), padding='same',
                     activation='relu', input_shape =input_size))
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(2, 2))
    model.add(Dropout(0.25))

    model.add(Conv2D(128, kernel_size=(3, 3), activation='relu',
                     padding='same', kernel_regularizer=regularizers.l2(0.01)))
    model.add(Conv2D(256, kernel_size=(3, 3), activation='relu',
                     kernel_regularizer=regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(1024, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(classes, activation='softmax'))

    #Compliling the model
    model.compile(optimizer=Adam(learning_rate=learning_rate),              # suggested start: 0.001, increase by 5
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Usage:
model_1 = model_1((row,col,1), num_classes)

ic(len(model_1.layers))
model_1.summary()
```

### **Define Model1 Callback Function**

```python
chk_path = 'model_1.h5'

log_dir = "checkpoint/logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

checkpoint = ModelCheckpoint(filepath=chk_path,
                             save_best_only=True,
                             verbose=1,
                             mode='min',
                             moniter='val_loss')

earlystop = EarlyStopping(monitor='val_loss',
                          min_delta=0,
                          patience=30,
                          verbose=1,
                          restore_best_weights=True)

early_stopping_2 = EarlyStopping(
                          patience=10,
                          min_delta=0.001,
                          monitor="val_loss",
                          restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(monitor='val_loss',
                              factor=0.2,
                              patience=6,
                              verbose=1,
                              min_delta=0.0001)


# tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
# csv_logger = CSVLogger('training.log')

# callbacks1 = [checkpoint, earlystop]
callbacks1 = [checkpoint, early_stopping_2]
ic(callbacks1);
```

```python
# Check Parameters
def check_parameters():
    ic(num_epochs)
    ic(steps_per_epoch)
    ic(validation_steps)
    ic(training_set.n // 1)
    ic(training_set.batch_size)
    ic(test_set.n // 1)
    ic(test_set.batch_size)

    ic(image_set)
    ic(num_classes)
    ic(learning_rate);

# Usage:
check_parameters()
```

```python
%%time
model1_hist = model_1.fit(x=training_set,
                   validation_data=val_set,
                   epochs=num_epochs,
                   callbacks=callbacks1,
                   steps_per_epoch=steps_per_epoch,
                   validation_steps=validation_steps)
```

## Model_2

### **Define/Compile Model2**

```python
def model_2(input_size, classes):
    model= keras.Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3),
                     padding='same', activation='relu',
                     input_shape=(48, 48,1)))
    model.add(Conv2D(64,(3,3), padding='same', activation='relu' ))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))


    model.add(Conv2D(128,(5,5), padding='same', activation='relu'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))


    model.add(Conv2D(512,(3,3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))


    model.add(Conv2D(512,(3,3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.01)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))

    #Faltten the model
    model.add(Flatten())


    model.add(Dense(256))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.25))


    model.add(Dense(512))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.25))

    model.add(Dense(classes, activation='softmax'))
    opt = Adam(learning_rate=learning_rate)
    model.compile(optimizer=opt,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
#    model.summary
    return model

# Usage:
model_2 = model_2((row,col,1), num_classes)

ic(len(model_2.layers))
model_2.summary()

```

## **VGG16 Model**

### **Importing the VGG16 Architecture**

* Not using fully connected layers of VVG16 model(include_top= False) as the VVG16 model has 1000 classes to my 4.
* Using pretrained weights of Imagenet

```python
vgg16_model = keras.applications.vgg16.VGG16(
    weights = 'imagenet',
    include_top = False,
    input_shape = (48, 48, 3))

for layer in vgg16_model.layers:
    layer.trainable = False
```

### **Build/Compile Model**

* Import VGG16 upto the the **'block5_pool'** layer
* Add layers: block5_pool => Flatten => Dense(256, relu) => Dense (Num_Classes, softmax)

```python
x = vgg16_model.get_layer('block5_pool').output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dense(512, activation='relu')(x)
x = Dense(num_classes, activation='softmax')(x)

vgg16_model = Model(inputs = vgg16_model.input, outputs = x)

vgg16_model.compile(
                loss = 'categorical_crossentropy',
                optimizer = 'Adam',
                metrics = ['accuracy'])

ic(len(vgg16_model.layers))

vgg16_model.summary()
```

### **Compile Checkpoints Functions**

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

chk_path = 'vgg16.h5'

log_dir = "checkpoint/logs/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

checkpoint = ModelCheckpoint(
                            filepath=chk_path,
                            save_best_only=True,
                            verbose=1,
                            mode='min',
                            moniter='val_loss')

early_stopping = EarlyStopping(
                            monitor = 'val_loss',
                            min_delta = 0,
                            patience = 10,
                            verbose = 1,
                            restore_best_weights = True)


reduce_learningrate = ReduceLROnPlateau(
                            monitor = 'val_loss',
                            factor = 0.2,
                            patience = 3,
                            verbose = 1,
                            min_delta = 0.0001)

#callbacks_list = [early_stopping, checkpoint]
callbacks_list = [checkpoint, early_stopping]
```

```python
# Check Parameters
def check_parameters():
    ic('== Parameters at runtime ==')
    ic(num_epochs)
    ic(steps_per_epoch)
    ic(validation_steps)
    ic(train_set.n // 1)
    ic(train_set.batch_size)
    ic(validation_set.n // 1)
    ic(validation_set.batch_size)

    ic(image_set)
    ic(num_classes)
    ic(data_classes)
    ic(callbacks_list[0])
    ic(learning_rate);

# Usage:
check_parameters()
```

### Fit/Train VGG16 Model

```python
%%time

vgg_hist = vgg16_model.fit(
            train_set,
            epochs = num_epochs,
            callbacks = callbacks_list,
            validation_data = validation_set)
```
